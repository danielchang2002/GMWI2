# Import libraries

```python
import pandas as pd
import numpy as np
import os
import config
from joblib import dump, load
```

```python
gmhi2 = load(os.path.join(config.MODEL_DIR, "gmhi2_model.joblib"))
gmhi2
```

# Load data

```python
X = pd.read_csv(os.path.join(config.CLEAN_DIR, "taxonomic_features_test.csv"), index_col=[0])
X.head()
```

# Compute scores and save

```python
output = pd.DataFrame(gmhi2.decision_function(X > config.PRESENCE_CUTOFF), index=X.index, columns=["GMHI2_score"])
output
```

```python
output.to_csv(os.path.join(config.LOG_DIR, "GMHI2_scores_test_set.csv"))
```
